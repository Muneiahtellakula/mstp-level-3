```python
import pandas as pd

data=pd.read_csv('olympics.csv',index_col=0, skiprows=1)
data
```

```python
for col in data.columns:
    if col[:2]=='01':
        data.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        data.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        data.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:2]=='№':
        data.rename(columns={col:'#'+col[4:]}, inplace=True)
#names_ids = data.index.str.split('\s\(')
names_ids = data.index.str.split('\s\(') # split the index by '('

data.index = names_ids.str[0] # the [0] element is the country name (new index) 
data['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

data = data.drop('Totals')
data.head()
```
